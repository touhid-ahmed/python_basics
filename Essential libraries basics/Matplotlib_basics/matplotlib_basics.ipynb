```python
import matplotlib.pyplot as plt
%matplotlib inline

heros_powers = {"Superman" : 5000, "Ironman": 3000, "Saitama":9000}

heros = list(heros_powers.keys())

powers = list(heros_powers.values())

fig = plt.figure(figsize=(3,5))

plt.bar(heros, powers, color = 'red', width=0.4)

plt.xlabel("Heroes")

plt.ylabel("Power Rating")

plt.title("Comparison of Superheros and their powers")

plt.show()
```
