```python
def tic_tac_toe():
    board_placement = {'A1' : ' ', 'A2' : ' ', 'A3' : ' ',
                       'B1' : ' ', 'B2' : ' ', 'B3' : ' ',
                       'C1' : ' ', 'C2' : ' ', 'C3' : ' '}
    
    def the_board(board):
        print(board['A1'] + "|" + board['A2'] + "|" + board['A3'] )
        print("-+-+-")
        print(board['B1'] + "|" + board['B2'] + "|" + board['B3'] )
        print("-+-+-")
        print(board['C1'] + "|" + board['C2'] + "|" + board['C3'])
    
    turn = 'X'
    while True:
    
        for i in range(9):
            the_board(board_placement)
            move = input("Turn for " + str(turn) + ". Move to which space?: ")
            
            if move == 'q':
                break
            
            else:
                board_placement[move] = turn
                
                if turn == 'X':
                    turn = '0'
                else:
                    turn = 'X'
                
#         if board_placement['A1'] and board_placement['A2'] and board_placement['A3'] == 'X':
#             print("X wins!")
#         elif board_placement['A1'] and board_placement['A2'] and board_placement['A3'] == '0':
#             print("0 wins!")
        the_board(board_placement)
        print("This is the final result")
        break

tic_tac_toe()
```
